```python
import requests
```

```python
# Set up authentication with access token
access_token = 'your_token'
headers = {'Authorization': f'token {access_token}'}

# Set up search query
query = '"are duplicates" in:body language:python state:closed label:refactoring type:pr'

# Set up endpoint for search API
endpoint = 'https://api.github.com/search/issues'

# Set up search parameters
params = {'q': query}

# Make API request
response = requests.get(endpoint, headers=headers, params=params)

# Parse response and get review messages
results = response.json()['items']
```

```python
class RenderHyperlink(object):
    def __init__(self, key, link, *args):
        link = link + "/" if not link.endswith("/") else link
        
        for arg in args:
            link += arg + "/"
            
        self.__url = '<a href={} target="_blank">{}</a>'.format(link, key)
    
    def __repr__(self):
        from IPython.display import display, HTML
        display(HTML(self.__url))
        return "" # hacky way to return a string despite not returning anything
```
